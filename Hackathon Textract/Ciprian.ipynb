### Get data from str-xml

```python
#Import modules
import pandas as pd
import xml.etree.ElementTree as ET
import io
import numpy as np
import os

#functie - parametrul xml_filename = numele fisierului xml
def xml2df():
    files_name = os.listdir(os.getcwd())
    
    final_list = []
    for file_name in files_name:
        final_list2 = []
        if("str.xml" in file_name):
            xml_data = open(file_name, encoding="utf8").read()
            root = ET.XML(xml_data) # element tree
            for table in root:
                table_coordinates = []
                for region in table:
                    page_number = region.get('page')
                    for cell in region:
                        coordinates = {}
                        for bbox in cell:
                            if(bbox.tag=="bounding-box"):
                                coordinates = {'x1':bbox.get('x1'), 'x2':bbox.get('x2'), 'y1':bbox.get('y1'), 'y2':bbox.get('y2'), 'page_number':page_number}
                            if(bbox.tag=="content"):
                                coordinates['content'] = bbox.text.strip()
                        table_coordinates.append(coordinates) 
                    final_list2.append(table_coordinates)

            for i in range(len(final_list2)):
                for j in range(len(final_list2[i])):
                    final_list2[i][j]['t_x1']=pd.DataFrame(final_list2[i])['x1'].min()
                    final_list2[i][j]['t_x2']=pd.DataFrame(final_list2[i])['x2'].max()
                    final_list2[i][j]['t_y1']=pd.DataFrame(final_list2[i])['y1'].min()
                    final_list2[i][j]['t_y2']=pd.DataFrame(final_list2[i])['y2'].max()
                    final_list2[i][j]['pdf_name']=file_name.strip('-str.xml')
                    
            final_list.append(final_list2)
    
    finalDataFrame = []    
    for array in final_list:
        for array2 in array:
            for array3 in array2:
                finalDataFrame.append(array3)
                   
    pd.DataFrame(finalDataFrame).to_csv("dataFromStrXml.csv", encoding='utf-8')   
    
xml2df()
```
